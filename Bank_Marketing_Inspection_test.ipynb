# Bank Marketing Data Analysis
This notebook presents an analysis of the `bankmarketing.csv` dataset, which includes data related to a bank's marketing campaigns. The main goal is to understand customer behavior and predict whether a client will subscribe to a term deposit.

```python

import pandas as pd

# Load the dataset
df = pd.read_csv('/bankmarketing.csv')

# Display the first few rows
df.head()
```

```python

# Check for missing values and data types
print(df.isnull().sum())
print("----------")
print(df.dtypes)
print("----------")
df.info()

```

 "The dataset contains 41,188 records with 21 columns. No missing values were found. Columns are a mix of numerical and categorical types."

```python

# Summary statistics
df.describe()

```

"The average customer age is 40 years, ranging from 17 to 98. This suggests a wide demographic reach in the bank's marketing campaign."

```python

import seaborn as sns
import matplotlib.pyplot as plt

# Plot the distribution of the target variable
sns.countplot(x='y', data=df)
plt.title('Customer Subscription Distribution')
plt.xlabel('Subscribed (yes/no)')
plt.ylabel('Number of Customers')
plt.show()
```

"88% of customers did not subscribe to a term deposit, indicating a heavily imbalanced dataset. This is important for future machine learning models."

```python

# Correlation matrix for numerical features

corr = df.select_dtypes(include='number').corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

"Most features show weak correlation with each other. Economic indicators like euribor3m and emp.var.rate show stronger relationships, which is expected."
